```python
from pathlib import Path

base = Path("../../data/raw_data/Segmentation/sama_coco_bag_seg")

def clean_split(split):
    img_dir = base / "images" / split
    lbl_dir = base / "labels" / split

    removed = 0
    for img in list(img_dir.iterdir()):
        if img.suffix.lower() not in [".jpg", ".jpeg", ".png"]:
            continue

        lab = lbl_dir / f"{img.stem}.txt"
        keep = True

        if not lab.exists():
            keep = False
        else:
            txt = lab.read_text().strip()
            if not txt:
                keep = False
            else:
                for line in txt.splitlines():
                    parts = line.split()
                    if len(parts) < 7 or (len(parts) - 1) % 2 != 0:
                        keep = False
                        break

        if not keep:
            img.unlink(missing_ok=True)
            lab.unlink(missing_ok=True)
            removed += 1

    print(f"{split}: removed {removed} bad samples")

clean_split("train")
clean_split("val")
```
